```python
import numpy as np
import matplotlib.pyplot as plt

# loading data
X = np.loadtxt('S.csv')
print(X[:10]) # check the first 10 entries
```

```python
def Confidence_Interval(mu, n, delta):
    epsilon = np.sqrt(1/(2*n)*np.log(2/(delta)))
    print(epsilon)
    lhs = np.max( [mu-epsilon, 0])
    rhs = np.min( [mu+epsilon, 1])
    return np.array([lhs, rhs])

mean = np.mean(X)
n = len(X)


print(Confidence_Interval(mean, n, 0.05))
print(Confidence_Interval(mean, n, 0.02))
```
