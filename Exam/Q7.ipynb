```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

data = np.loadtxt('Q7-dataset.txt', delimiter=",")
halfway = len(data) // 2

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')
ax.scatter(data[:halfway, 0], data[:halfway, 1], data[:halfway, 2], c='blue', label='First half')
ax.scatter(data[halfway:, 0], data[halfway:, 1], data[halfway:, 2], c='red', label='Second half')
ax.set_title("Scatterplot of Q7-dataset.txt")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
plt.legend()
plt.show()
```

## 7.2

```python
# only keep the first 2 PC to reduce to 2D
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data)

plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:halfway, 0], data_pca[:halfway, 1], c='blue', label='First half')
plt.scatter(data_pca[halfway:, 0], data_pca[halfway:, 1], c='red', label='Second half')
plt.title("PCA Plot of Q7-dataset.txt")
plt.xlabel("Principle component 1")
plt.ylabel("Principle component 2")
plt.legend()
plt.show()
```

## 7.3

```python
perplexities = [30, 100, 300]
fig, axes = plt.subplots(1, 3, figsize=(16, 6))

for i, perplexity in enumerate(perplexities):
    tsne = TSNE(perplexity=perplexity)
    data_tsne = tsne.fit_transform(data)
    
    axes[i].scatter(data_tsne[:halfway, 0], data_tsne[:halfway, 1], c='blue', label='First half')
    axes[i].scatter(data_tsne[halfway:, 0], data_tsne[halfway:, 1], c='red', label='Second half')
    axes[i].set_title(f"t-SNE with perplexity={perplexity}")
    axes[i].set_xlabel("t-SNE 1")
    axes[i].set_ylabel("t-SNE 2")

plt.legend()
plt.show()
```
