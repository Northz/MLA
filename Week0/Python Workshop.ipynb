```python
# get first two columns for three species
setosa_feature = []
versicolor_features = []
veirginica_features = []

with open("iris.data", "r") as file:
    all_rows = csv.reader(file)
    for row in all_rows:
#         print(row)
        if row:
            if row[-1] == "Iris-setosa":
                setosa_feature.append([float(row[0]), float(row[1])])
            if row[-1] == "Iris-versicolor":
                versicolor_features.append([float(row[0]), float(row[1])])
            if row[-1] == "Iris-virginica":
                veirginica_features.append([float(row[0]), float(row[1])])

print(len(setosa_feature), len(versicolor_features), len(veirginica_features))
```

```python
plt.figure(figsize=(10, 6))
plt.scatter([point[0] for point in setosa_feature], [point[1] for point in setosa_feature], label="Setosa", c="red")
plt.scatter([point[0] for point in versicolor_features], [point[1] for point in versicolor_features], label="Versicolor")
plt.scatter([point[0] for point in veirginica_features], [point[1] for point in veirginica_features], label="Verginica")
plt.legend(title="Sepcies")
plt.title("Scatter plot")
plt.xlabel("Sepal Legth")
plt.ylabel("Sepal Width")
plt.savefig("scatterplot.png")
plt.show()
```

```python
def mean_value(feature):
    return sum(feature)/ len(feature)
```

```python
s_legth_setosa = [point[0] for point in setosa_feature]
s_legth_avg= mean_value(s_legth_setosa)
print("Average sepal legth: ", s_legth_avg)
print("Average sepal legth using NumPy: ", np.mean(s_legth_setosa))
```

```python
s_legth_versicolor = [point[0] for point in versicolor_features]
s_legth_avg= mean_value(s_legth_versicolor)
print("Average sepal legth: ", s_legth_avg)
print("Average sepal legth using NumPy: ", np.mean(s_legth_versicolor))
```

```python
s_legth_veirginica = [point[0] for point in veirginica_features]
s_legth_avg= mean_value(s_legth_veirginica)
print("Average sepal legth: ", s_legth_avg)
print("Average sepal legth using NumPy: ", np.mean(s_legth_veirginica))
```

# Question-4

```python
# feature = [sl1, sl2, sl3....]
def variance(feature):
    mean_val = sum(feature)/ len(feature)
    squared_diff = [(point-mean_val)**2 for point in feature]
    variance_ = sum(squared_diff)/len(feature)
    return variance_
```

```python
print("Variance of Sepal leght of Setosa using NumPy", np.var(s_legth_setosa))
print("Variance of Sepal leght of Setosa", variance(s_legth_setosa))
```

```python
print("Variance of Sepal leght of Setosa using NumPy", np.var(s_legth_versicolor))
print("Variance of Sepal leght of Setosa", variance(s_legth_versicolor))
```

```python
print("Variance of Sepal leght of Setosa using NumPy", np.var(s_legth_veirginica))
print("Variance of Sepal leght of Setosa", variance(s_legth_veirginica))
```

```python
# create data arry for entire
data = []
with open("iris.data", "r") as file:
    all_rows = csv.reader(file)
    for row in all_rows:
        if row:
            data.append([float(row[0]), float(row[1]), float(row[2]), float(row[3])])
# print(data)
```

```python
# Euclidian distance between two points without using built-in functions
first_point = data[0]
second_point = data[1]
print(first_point, second_point)
```

```python
dist_bw_two_points = sum([(point1 - point2)**2 for point1, point2 in zip(first_point, second_point)])**0.5
print(dist_bw_two_points)
```

```python
# satrt using builtin functions
data_arr = np.array(data)
data_arr.shape
```

```python
print(data_arr[0].shape, data_arr[1:].shape)
```

```python
first_point = data_arr[0]
remaining_points = data_arr[1:]
distances = np.sqrt(np.sum((remaining_points - first_point)**2, axis=1))
print(distances)
```

```python
print("average distance", np.mean(distances))
print("variance of distances", np.var(distances))
print("varaince of distances using custom", variance(distances))
```
